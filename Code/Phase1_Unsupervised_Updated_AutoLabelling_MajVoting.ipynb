## Function for Text Cleaning

```python
def data_cleaner(data):
    clean_data = []
    lemmatizer = WordNetLemmatizer()
    stopword = set(stopwords.words('english'))
    custom_stopwords = ["okay", "uh"]
    
    for sentence in tqdm(data):
        # Step 1: Clean HTML tags, remove content within square brackets, remove characters except letters
        cleantext = re.sub('\[[^]]*\]', ' ', BeautifulSoup(sentence, "lxml").text)
        cleantext = re.sub('[^a-zA-Z]', ' ', cleantext)

        # Step 2: Remove punctuation
        cleantext = re.sub(r'[^\w\s]', '', cleantext)

        # Step 3: Convert to lowercase, Tokenize the text and remove stopwords and custom stopwords
        tokens = word_tokenize(cleantext.lower())
        tokens_cleaned = [token for token in tokens if token not in stopword and token not in custom_stopwords and len(token) > 1]

        # Step 4: Lemmatize the tokens
        lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens_cleaned]

        # Step 5: Remove any leading or trailing whitespace
        cleantext = ' '.join(lemmatized_tokens).strip()
        
        clean_data.append(cleantext)
    
    return clean_data
```

```python
#custom_stopwords = ["okay", "uh"]  # Add extra stopwords here
data['Overall_Transcript_clean'] = data_cleaner(data['Transcribe_output'])
data['Agent_Transcript_clean'] = data_cleaner(data['AgentTranscription'])
data['Cust_Transcript_clean'] = data_cleaner(data['CustomerTranscription'])
```

# Calculating Sentiment Polarity using SentiWordNet

**Note:** When working with SentiWordNet to determine the sentiment of words, it is necessary to know the characteristic of the word for which we want to know the sentimet. So for identify the position of the word we need, part of speech (POS) tags for each word in the sentence. NLTK's POS tagging functionality is useful in this regard, as it assigns POS tags to words based on their contextual usage.

1. NLTK uses the following POS tags for word classification:

- 'NN' for NOUN
- 'VB' for VERB
- 'JJ' for ADJECTIVE
- 'ADV' for ADVERB

These POS tags are crucial in retrieving the appropriate sentiment scores from SentiWordNet. By mapping the POS tags obtained through NLTK to the compatible POS tags of SentiWordNet, it becomes possible to obtain sentiment scores for words. 
The positive and negative scores can then be averaged to determine the overall sentiment of a sentence or text passage.

- [Penn Tree Bank POS Tags List](https://www.ling.upenn.edu/courses/Fall_2007/ling001/penn_treebank_pos.html)
- [NLTK POS Tags List](https://www.learntek.org/blog/categorizing-pos-tagging-nltk-python/)
- [WordNet POS Tag List](https://www.geeksforgeeks.org/nlp-wordnet-for-tagging/)

```python
# Perform POS tagging on pre-processed data
def perform_pos_tagging(column):
    postagging = []
    for review in tqdm(column):
        tokens = word_tokenize(review)
        postagging.append(nltk.pos_tag(tokens))
    return postagging
```

## PennTreebank tags to WordNet tags

**1. PennTreebank POS Tags:**

- The PennTreebank POS tag set is a widely used set of tags for part-of-speech (POS) annotation of words in natural language processing tasks.
- It provides a standard way to categorize words into different syntactic categories based on their grammatical roles and behavior in a sentence.
- Examples of PennTreebank POS tags include 'NN' (singular noun), 'VBG' (gerund or present participle), 'JJ' (adjective), 'RB' (adverb), etc.
- These tags capture important linguistic information about words, allowing for more precise analysis and processing of text.

**2. WordNet POS Tags:**

- WordNet is a lexical database that organizes words into sets of synonyms called synsets and provides semantic relationships between them.
- WordNet uses its own set of POS tags to classify words based on their lexical categories and meanings.
- WordNet POS tags include 'n' (noun), 'v' (verb), 'a' (adjective), 's' (adjective satellite), and 'r' (adverb).
- These tags are used to disambiguate word senses, retrieve relevant synsets, and perform sentiment analysis and other language-related tasks.

**Importance of the penn_to_wn function:**

- The penn_to_wn function is important because it allows us to map PennTreebank POS tags to WordNet POS tags.
- Since different libraries and resources use different tag sets, this mapping helps bridge the gap between them.
- In sentiment analysis, for example, we need to identify the POS tags of words to determine their semantic orientation (positive, negative, or neutral).
- By converting PennTreebank tags to WordNet tags, we can then retrieve the sentiment scores from resources like SentiWordNet, which associate scores with specific WordNet synsets.
- The penn_to_wn function enables this conversion, making it easier to leverage the sentiment information available in resources like SentiWordNet.

```python
# Map PennTreebank tags to WordNet tags, Taking only relavant tags for sentiment analysis.

def penn_to_wn(tag):
    if tag is None:
        return None
    if tag.startswith('J'):    # Adjective
        return wn.ADJ
    elif tag.startswith('N'):  # Noun
        return wn.NOUN
    elif tag.startswith('R'):  # Adverb
        return wn.ADV
    elif tag.startswith('V'):  # Verb
        return wn.VERB
    return None
```

```python
# Apply penn_to_wn function on pos_tags
wn_tags = [(token, penn_to_wn(tag)) for token, tag in pos_tags]

# Print the WordNet tags
print(wn_tags)
```

```python
# Get sentiment score for a word based on its POS tag
def get_sentiment(word, tag):
    # Map PennTreebank tag to WordNet tag
    wn_tag = penn_to_wn(tag)
    
    # Check if the WordNet tag is relevant for sentiment analysis
    if wn_tag not in (wn.NOUN, wn.ADJ, wn.ADV):                                 #wn.verb
        return []

    # Lemmatization
    lemma = lemmatizer.lemmatize(word, pos=wn_tag)
    if not lemma:
        return []

    #Synset is a special kind of a simple interface that is present in NLTK to look up words in WordNet. 
    #Synset instances are the groupings of synonymous words that express the same concept. 
    #Some of the words have only one Synset and some have several
    
    # Look up synsets in WordNet
    synsets = wn.synsets(word, pos=wn_tag)
    if not synsets:
        return []

    # Take the first sense, the most common
    synset = synsets[0]
    
    # Get sentiment scores from SentiWordNet for the synset
    swn_synset = swn.senti_synset(synset.name())
    
    # Return a list containing the synset name, positive score, negative score, and objective score
    return [synset.name(), swn_synset.pos_score(), swn_synset.neg_score(), swn_synset.obj_score()]
```

```python
# Calculate sentiment score based on POS tags in a column
def calculate_sentiment_score(column, pos_column):
    # Initialize variables
    pos = neg = obj = count = 0
    
    # Calculate sentiment score for each review
    senti_score = []
    for pos_val in column:
        senti_val = [get_sentiment(x, y) for (x, y) in pos_val if y is not None]
        for score in senti_val:
            try:
                pos += score[1]  # positive score is stored at 2nd position
                neg += score[2]  # negative score is stored at 3rd position
            except:
                continue
        senti_score.append(pos - neg)
        pos = neg = 0
    
    # Assign sentiment scores to the 'senti_score' column in the DataFrame
    data[pos_column] = senti_score
```

```python
# Calculate sentiment score for 'pos_tags_overall' column and assign scores to 'senti_score_overall' column
calculate_sentiment_score(data['pos_tags_overall'], 'senti_score_overall')

# Calculate sentiment score for 'pos_tags_agent' column and assign scores to 'senti_score_agent' column
calculate_sentiment_score(data['pos_tags_agent'], 'senti_score_agent')

# Calculate sentiment score for 'pos_tags_customer' column and assign scores to 'senti_score_customer' column
calculate_sentiment_score(data['pos_tags_customer'], 'senti_score_customer')
```
